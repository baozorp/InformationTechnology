```python
import plotly.express as px
import numpy as np
import math
import sympy

def f(x):
  return x**4-18*x**3+6

x = np.arange(-5.0, 5.0, 0.001)

fig = px.scatter(x=x, y=f(x))

fig = px.scatter(x=x, y=f(x))
fig.update_xaxes(range = [x[0], x[-1]], zeroline=True, zerolinewidth=5, zerolinecolor='Green')
fig.update_yaxes(range=[-10.0, 10.0], zeroline=True, zerolinewidth=5, zerolinecolor='Green')

fig.show()
```

```python
a = -2.0; b = 2.0; e = 0.001

def f(x):
    return x**4-18*x**3+6


if f(a) * f(b) >= 0:
    print("Корень не найден") 

else:
    n = 1
    x = (a+b)/2
    while (abs(a-b) > e):
        x = (a+b)/2

        if f(a) * f(x) < 0:
            b = x
        else:
            a = x
        n += 1
    print (f"Корень уравнения = {x}, количество шагов = {n}")
```
